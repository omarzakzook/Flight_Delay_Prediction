## Project Overview

Flight delays create operational costs for airlines and airports and cause significant inconvenience for passengers.

This project develops and evaluates machine learning models for predicting whether a completed U.S. domestic flight will arrive **15 minutes or more late**.

The project initially evaluates a pre-departure prediction scenario using scheduled flight, airline, airport, calendar, and seasonal information. Because these variables provide limited predictive signal, a second operational scenario is evaluated using information available immediately after departure.

The final selected system is an **after-departure arrival-delay prediction model** because it uses actual departure time and departure delay.

# ❓ 2. Problem Statement

Flight delays affect airline operations, airport resource planning, passenger connections, and customer satisfaction.

The original business objective was to predict arrival delays before departure. However, the project experiments demonstrated that the available pre-departure variables provided limited predictive performance.

The final machine learning problem is therefore defined as:

> Immediately after a flight departs, predict whether it will arrive at least 15 minutes later than its scheduled arrival time.

At this prediction moment, the model can use:

- Scheduled flight information
- Airline and airport information
- Calendar and seasonal information
- Actual departure time
- Departure delay

The system does not use arrival outcomes, elapsed-time outcomes, delay-cause variables, or other information that occurs after the prediction moment.

```python
IS_DELAYED = (ARR_DELAY >= 15).astype(int)

# 🛠️ 8. Project Methodology

The project follows a structured CRISP-DM-inspired workflow:

1. Business Understanding
2. Data Understanding
3. Data Cleaning
4. Feature Engineering
5. Exploratory Data Analysis
6. Data Preparation
7. Baseline Model Comparison
8. Temporal Hyperparameter Tuning
9. Prediction-Scenario Improvement
10. Final Temporal Holdout Evaluation
11. Model Persistence and Deployment

The project uses chronological periods rather than a random split:

- Training: 2019–2021
- Validation: 2022
- Temporal holdout test: January–August 2023

The temporal holdout is not used for fitting, hyperparameter tuning, threshold selection, or model comparison.

# 📚 9. Project Roadmap

| Notebook | Purpose |
|---|---|
| 01 | Business Understanding |
| 02 | Data Understanding |
| 03 | Data Cleaning and Target Creation |
| 04 | Feature Engineering and Leakage Control |
| 05 | Exploratory Data Analysis Using the Training Period |
| 06 | Data Preparation and Preprocessing Pipeline |
| 07 | Baseline Model Comparison and Temporal Hyperparameter Tuning |
| 08 | After-Departure Model Improvement |
| 09 | Final Training, Temporal Holdout Evaluation, and Artifact Saving |

The deployed Streamlit application loads the final fitted pipeline and deployment metadata saved by Notebook 9.

# ⚠️ 10. Scope and Limitations

The project has several important limitations:

- The final model predicts only after the aircraft has departed.
- The modeling population excludes cancelled and diverted flights.
- The 2023 data ends in August and does not represent a complete calendar year.
- The dataset does not include detailed weather forecasts, air-traffic congestion, maintenance events, or real-time airport conditions.
- The delayed-class model score is not calibrated as a guaranteed real-world probability.
- Aviation patterns can change over time, so the model should be monitored for temporal drift.
- The 2020 pandemic period has a different operational distribution from later years.
- Predictions outside the 2019–2023 development period should be interpreted cautiously.
